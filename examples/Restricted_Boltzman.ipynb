```python
import pandas as pd
from sklearn.neural_network import BernoulliRBM
import matplotlib.pyplot as plt
import random as rd
%matplotlib inline
```

```python
recipes = pd.DataFrame.from_csv('ReceptenBinair_minimun2ingredients.csv', sep=';')
recipes
```

```python
ingredienten = recipes.columns
print ingredienten
```

```python
rbm = BernoulliRBM(n_components=50, verbose=True, learning_rate=0.05, batch_size=100, n_iter=25)
rbm.fit(recipes)
```

```python
def bin_to_recipe(binary_vector):
    return [ingr for (ingr, ind) in zip(ingredienten, binary_vector) if ind == 1]

def recipe_to_bin(recipe):
    return [1 if ingr in recipe else 0 for ingr in ingredienten]
```

```python
# 10 verbeteringen:

for i in rd.sample(range(50000), 10):
    rec = recipes.values[i]
    changed_rec = rbm.gibbs(rec)
    print ', '.join(bin_to_recipe(rec)), '=>', ', '.join(bin_to_recipe(changed_rec))
    print 
```

```python
# pseudo likelihood

good_recipe = ['hazelnut', 'vanilla', 'milk', 'cocoa']
bad_recipe = good_recipe + ['garlic']

print 'Pseudolikelihood good recipe', rbm.score_samples(recipe_to_bin(good_recipe))
print 'Pseudolikelihood bad recipe', rbm.score_samples(recipe_to_bin(bad_recipe))
```
